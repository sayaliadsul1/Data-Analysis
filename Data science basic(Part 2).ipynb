# 3. iloc = It is based on indexing used for slicing the data

# a. select rows with indices using iloc

```python
heart.iloc[[0,2]]
```

```python
heart.iloc[[3,9]]
```

```python
heart.iloc[[900,101]]
```

# b. select rows with particular indices & columns

```python
h1 = heart.iloc[[0,2],[1,3]]
print("1st square bracket [0,2] indicates rows numbers")
print("2nd square bracket [1,3] indicates column numbers")
print("-------------------------------------------------------")
print(h1)
```

```python
h2 = heart.iloc[[4,5],[3,4]]
print("1st square bracket [4,5] indicates rows numbers")
print("2nd square bracket [3,4] indicates column numbers")
print("-------------------------------------------------------")
print(h2)
```

```python
h3 = heart.iloc[[300,400],[3,4]]
print("1st square bracket [300,400] indicates rows numbers")
print("2nd square bracket [3,4] indicates column numbers")
print("-------------------------------------------------------")
print(h3)
```

# c. select a range of rows using iloc

```python
a1 = heart.iloc[0:3]
print("slice the data 0 to 3 rows")
print("------------------------------")
a1
```

```python
a2 = heart.iloc[0:50]
print("slice the data 0 to 50 rows")
print("------------------------------")
a2
```

# d. select a range of rows & columns using iloc

```python
b1 = heart.iloc[0:20,2:6]
print("slice the data 0 to 20 rows && 2 to 6 columns")
print("----------------------------------------------")
b1
```

```python
b2 = heart.iloc[0:10,0:5]
print("slice the data 0 to 10 rows && 0 to 5 columns")
print("----------------------------------------------")
b2
```
