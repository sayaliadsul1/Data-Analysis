```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv(r"C:\Users\Acer\Desktop\Dataset\heart_disease dataset.csv")
```

```python
data
```

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
from sklearn import metrics
rmse = np.sqrt(metrics.mean_squared_error(y_test,y_pred))
rmse
```

### See that without normalization the root mean squared error value comes out to be 0.3950,

# Using Standardization

```python
data
```

```python
x1=data.iloc[:,:-1].values
y1=data.iloc[:,-1].values
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(x1)
x1= sc.transform(x1)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x1,y1,test_size=0.2)

```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y1_pred=lr.predict(x_test)
y1_pred
```

```python
from sklearn import metrics
rmse = np.sqrt(metrics.mean_squared_error(y_test,y1_pred))
rmse
```

### See that, we are able to get a significant reduction in the error when we used the standardization technique.
