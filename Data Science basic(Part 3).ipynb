# loc = It is based on conditions for slicing the data

# a. all th rows based on any condition in a column

```python
## find out the age is less than equal to 40
```

```python
age1 = heart.loc[heart.Age<=40]
age1
```

```python
age1["Age"].min()
```

```python
age1["Age"].max()
```

```python
## find the BP rate is 140
```

```python
bp = heart.loc[heart.RestingBP==140]
bp
```

```python
bp["RestingBP"].min()
```

```python
bp["RestingBP"].max()
```

# b. find all the rows with more than one condition

```python
## find out male age is less than 30
```

```python
h1 = heart[['Age','Sex']]
h1
```

```python
h2 = h1.loc[(h1.Age<=30) & (h1.Sex=='M')]
h2
```
